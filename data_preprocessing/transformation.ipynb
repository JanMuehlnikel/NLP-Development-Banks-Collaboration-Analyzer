# **Data Transformation**

## **Dependecies**

```python
import pandas as pd
import ast
```

## **Helper**

```python
def parse_list(text):
    return ast.literal_eval(text)
```

## **Load CSV**

Load the csv fetched by the api

```python
df = pd.read_csv("../data_collection/data/response_iati.csv")

print(df.info())   
df.head(10)
```

## **Create new data frame**

```python
trans_df = pd.DataFrame()
```

## **Transfrom columns**

### Transform ID

```python
trans_df['id'] = df['iati_identifier']
trans_df.head(2)
```

### Transform title

It was decided to use the english title, if available, preferred to other languages. This was the case to enhance the comparability in the later stages of this analysis.

```python
trans_df['title'] = df['title_narrative'].apply(lambda x: parse_list(x)[0])

# DECIDED TO USE ENGLISH TITLE PREFERRED TO OTHER LANGUAGES
for index, row in df.iterrows():
    try:
        # list of all available languages for the title
        lang_list = parse_list(row['title_narrative_xml_lang'])

        # if english is available as language find out the index and use the english version of the title
        if "en" in lang_list:
            en_idx = lang_list.index("en")
                
            trans_df.title[index] = parse_list(df.title_narrative[index])[en_idx]
    except:
        pass

trans_df.head(10)
```

### Transform organization

```python
trans_df['organization'] = df['reporting_org_narrative'].apply(lambda x: parse_list(x)[0])
trans_df.head(2)
```

### Transform description

```python
trans_df['description'] = df['description_narrative'].apply(lambda x: parse_list(x)[0])

# DECIDED TO USE ENGLISH TITLE PREFERRED TO OTHER LANGUAGES
for index, row in df.iterrows():
    try:
        # list of all available languages for the title
        lang_list = parse_list(row['title_narrative_xml_lang'])

        # if english is available as language find out the index and use the english version of the title
        if "en" in lang_list:
            en_idx = lang_list.index("en")
                
            trans_df.title[index] = parse_list(df.title_narrative[index])[en_idx]
        else:
            pass
    except:
        pass

trans_df.head(10)
```
