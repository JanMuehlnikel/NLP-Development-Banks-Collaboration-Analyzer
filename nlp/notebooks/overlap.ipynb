## **Config**

```python
df = pd.read_csv("../../src/merged_orgas.csv")

pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)  # Auto-detect the display width
pd.set_option('display.max_colwidth', None)  # Show full content of each column

df.head(1)
```

## **CRS3 & Country Overlap**

```python
filtered_country_df = df[df["crs_3_code"].str.contains('313', na=False)]
filtered_df = filtered_country_df[filtered_country_df["country"].str.contains('PH', na=False)]

filtered_df[["organization", "title_main", "description_main", "country", "region", "crs_3_code", "crs_3_name"]]
```
